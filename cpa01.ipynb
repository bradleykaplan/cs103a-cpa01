# Description:
### The dataset is B Corporation certification scores. B Corporation certifies different companies based on social and environmental performance categories, essentially having a business structure that is accountable to the environment and other social needs, as well as shareholders. The companies are scored on this criteria and, among other things, need an overall score of at least 80 to be certified. More info: https://www.bcorporation.net/en-us/certification

```python
import numpy as np
from matplotlib import pyplot as plt
#graph of overall score for all certifications (may contain duplicate certifications of the same company and de-certified companies)
column = "overall_score"
# used this to get data type: https://www.w3schools.com/python/gloss_python_getting_data_type.asp
# print(type(bcorps[column]))
# learn more about Series data type: https://pandas.pydata.org/docs/reference/api/pandas.Series.html?highlight=series#pandas.Series
pd.Series(sorted(bcorps[column])).plot(figsize=(15,10))
plt.grid()
plt.title(column,fontsize=32)
```

```python
#want to clean data set to only have currently certified B Corporations, and of those only the most recent certification information
#used sort_values info: https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.sort_values.html
#used drop duplicates info: https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.drop_duplicates.html
cbcorps = bcorps.sort_values(by=['company_id', 'assessment_year'])
cbcorps = cbcorps.drop_duplicates(subset = ['company_id'], keep = 'last')
cbcorps
```

```python
count = 0
for row in cbcorps.iterrows():
    if dict(list(row)[1])['current_status'] != 'certified':
        count += 1
print(count)
```

```python
#used iterrows info: https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.iterrows.html#pandas.DataFrame.iterrows
#used tuple to list conversion: https://pythonguides.com/python-convert-tuple-to-list/
#used dict conversion: https://www.tutorialspoint.com/How-I-can-convert-a-Python-Tuple-into-Dictionary
for row in cbcorps.iterrows():
    if dict(list(row)[1])['current_status'] != 'certified':
        droplist = []
        droplist.append(list(row)[0])
        cbcorps = cbcorps.drop(droplist)
```
